```python
#import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in file
file_path = "Resources/Reason_for_assistance.csv"
```

```python
# Read and display the CSV with Pandas
summary_df = pd.read_csv(file_path,encoding='ISO-8859-1')
summary_df
```

```python
#Clean dataframe
#Changing the name of columns/headers

summary_df = summary_df.rename(columns={
    'Table of contents': 'State/territory', 
    'Unnamed: 1': 'Group',
    'Unnamed: 2':'Reason for seeking assistance',
    'Unnamed: 3':'Males',
    'Unnamed: 4':'Females',
    'Unnamed: 5':'Total clients (number)',
    'Unnamed: 6':'Total clients (per cent)'
})

#Deleting first two rows (irrelavant)
summary_df = summary_df.iloc[2:]
#drop NA
summary_df = summary_df.dropna(how = "any")
#filtering only WA data
clean_df = summary_df[summary_df['State/territory'] == "WA"] 
#removing commas from Total clients (number)

clean_df["Total clients (number)"] = clean_df["Total clients (number)"].str.replace(",","")
clean_df["Males"] = clean_df["Males"].str.replace(",","")
clean_df["Females"] = clean_df["Females"].str.replace(",","")


clean_df = clean_df.astype({"Total clients (number)":"float"})
clean_df = clean_df.astype({"Males":"float"})
clean_df = clean_df.astype({"Females":"float"})


clean_df

```

```python
#Using Groupby

grouped_summary_df = clean_df.groupby(['Group']).sum()
grouped_summary_df
```

```python
# Initial assumption - based on personal encounter/observation

# (assumption 1) Males are more subjected to homelessness
# (assumption 2) Drug and Alcohol would be the main reason for homeless as most homeless are under the influence
```

# Gender Analysis to answer (assumption 1)


```python
# Gender Analysis to answer (assumption 1)

Male_stat = clean_df["Males"].sum()
Female_stat = clean_df["Females"].sum()
Ratio_stat = (Female_stat/Male_stat)
Ratio_stat = round(Ratio_stat,2)
print(f"The total number of male homeless in WA is: {Male_stat}")
print(f"The total number of female homeless in WA is: {Female_stat}")
print(f"Females are {Ratio_stat} times are more likely to be homeless than males")
```

# Visualizing top contributors of homelessness by group

```python
# Creating a pie chart two seperate groups

# Labels for the sections of our pie chart
labels = ["Accommodation", "Financial", "Health", "Interpersonal relationships", "Other"]
sizes = [21968.0, 28044.0, 15893.084, 34441.0, 24202.0]

# The colors of each section of the pie chart
colors = ["lightskyblue", "orange", "lightcoral", "red", "yellow"]

# Tells matplotlib to seperate the "Humans" section from the others
explode = (0, 0, 0, 0.1,0)
```

```python
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Give our chart some labels and a tile
plt.title("Main contributing factors to homelessness")
plt.show()
```

# Validating (assumption 2) Drug and Alcohol would be the main contributing factors to homelessness as most homeless are under the influence


```python
# Organising data frame starting from most to least number of clients


clean_df[["Reason for seeking assistance", "Total clients (number)"]].groupby("Reason for seeking assistance").max()
clean_df = clean_df.sort_values("Total clients (number)",ascending=False)
clean_df.head()
```

# Analysis - top contributors of homelessness

- Interpersonal Relationship issues are the main reason for clients to be homelessness which represents 27.7% of the whole pool of candidates.
- Top 3 contributors total client figures (percentage) are very close to each other.
- Intpersonal Relationship (27%), Financial Reason (22.5%) and Other/miscellaneous (19.4%)
- Drug and Subtance abuse are insignficant and does not form part of the top 5 contributing factors.

```python
#Obtain top 5 reasons for homelessness

Group_reasons_df = clean_df[["Reason for seeking assistance","Males","Females"]].head(5)
Group_reasons_df

Reason_for_seeking_assistance = Group_reasons_df["Reason for seeking assistance"]
Male_client_num = Group_reasons_df["Males"]
Female_client_num = Group_reasons_df["Females"]

x = np.arange(len(Reason_for_seeking_assistance))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Male_client_num, width, label='Men')
rects2 = ax.bar(x + width/2, Female_client_num, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total Client Number')
ax.set_title('Top reasons for Homelessness by gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
plt.xticks(rotation=90)

plt.ylim((0,12000))


plt.show()
```

# Analysis - Top reasons for Homelessness by Gender

- There are more female clients compared to male.
- There were more recorded female clients in every categories (i.e. reasons for becoming homeless).
- The biggest imbalance between the two genders are Interpersonal Relationships and accomodation.
- Female clients were 3.43 times more affectrf by interpersonal relationships followed by - Accommodation, 2.44 times.
- Gender Ratio per reasons (Female (1): Male)
