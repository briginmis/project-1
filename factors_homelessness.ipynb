```python
#import dependencies

# %matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load in file
file_path = "Resources/Reason_for_assistance.csv"
```

```python
# Read and display the CSV with Pandas
summary_df = pd.read_csv(file_path,encoding='ISO-8859-1')
summary_df
```

```python
#Clean dataframe
#Changing the name of columns/headers

summary_df = summary_df.rename(columns={
    'Table of contents': 'State/territory', 
    'Unnamed: 1': 'Group',
    'Unnamed: 2':'Reason for seeking assistance',
    'Unnamed: 3':'Males',
    'Unnamed: 4':'Females',
    'Unnamed: 5':'Total clients (number)',
    'Unnamed: 6':'Total clients (per cent)'
})

#Deleting first two rows (irrelavant)
summary_df = summary_df.iloc[2:]
#drop NA
summary_df = summary_df.dropna(how = "any")
#filtering only WA data
clean_df = summary_df[summary_df['State/territory'] == "WA"] 
#removing commas from Total clients (number)

clean_df["Total clients (number)"] = clean_df["Total clients (number)"].str.replace(",","")
clean_df["Males"] = clean_df["Males"].str.replace(",","")
clean_df["Females"] = clean_df["Females"].str.replace(",","")


clean_df = clean_df.astype({"Total clients (number)":"float"})
clean_df = clean_df.astype({"Males":"float"})
clean_df = clean_df.astype({"Females":"float"})


clean_df

```

```python
#Using Groupby

grouped_summary_df = clean_df.groupby(['Group']).sum()
grouped_summary_df
```

```python
# Initial assumption - based on personal encounter/observation

# (assumption 1) Males are more subjected to homeless
# (assumption 2) Drug and Alcohol would be the main reason for homeless as most homeless are under the influence
```

# Gender Analysis to answer (assumption 1)


```python
# Gender Analysis to answer (assumption 1)

Male_stat = clean_df["Males"].sum()
Female_stat = clean_df["Females"].sum()
Ratio_stat = (Female_stat/Male_stat)
Ratio_stat = round(Ratio_stat,2)
print(f"The total number of male homeless in WA is: {Male_stat}")
print(f"The total number of female homeless in WA is: {Female_stat}")
print(f"Females are {Ratio_stat} times are more likely to be homeless than males")
```

# Family and Domestic violence analysis

```python
# Load in file
file_path = "Resources/domestic_violence.csv"
```

```python
# Read and display the CSV with Pandas
d_violence_df = pd.read_csv(file_path,encoding='ISO-8859-1')
d_violence_df
```

```python
#Clean dataframe
#Changing the name of columns/headers

d_violence_df = d_violence_df.rename(columns={
    'Table of contents': 'Data type', 
    'Unnamed: 1': 'Age group',
    'Unnamed: 2':'Males',
    'Unnamed: 3':'Males Percent',
    'Unnamed: 4':'Females',
    'Unnamed: 5':'Females Percent',
    'Unnamed: 6':'All clients',
    'Unnamed: 7':'All clients (percent)'

})

#Deleting first 4 rows (irrelavant)
d_violence_df = d_violence_df.iloc[4:]
# d_violence_df = d_violence_df.drop(d_violence_df.index[14:29])

#drop NA
d_violence_df = d_violence_df.dropna(how = "any")

# #removing commas from males and females

d_violence_df["Males"] = d_violence_df["Males"].str.replace(",","")
d_violence_df["Females"] = d_violence_df["Females"].str.replace(",","")


d_violence_df = d_violence_df.astype({"Males":"float"})
d_violence_df = d_violence_df.astype({"Females":"float"})


# 
d_violence_df
```

```python
d_violence_df.drop(d_violence_df.loc[13:23].index, inplace=True)
d_violence_df
```

```python
d_violence_df = d_violence_df[["Age group","Males","Females"]]
d_violence_df
```

```python
Age_group = d_violence_df["Age group"]
male_v = d_violence_df["Males"]
female_v = d_violence_df["Females"]


Male, = plt.plot(Age_group, male_v, color="green", label="Male affected" )
Female, = plt.plot(Age_group, female_v, color="red", label="female affected" )

# Create a legend for our chart
plt.legend(handles=[Male, Female], loc="best")

# Resetting axis limits on my plot
# ax.set_ylim([100, 14000])# Show the chart
# plt.ylim(100.0, 25000.0)
plt.title("Client affected by Domestic Violence in Australia")
plt.xlabel("Age Group")
plt.ylabel("Client number")

plt.show()
```
